# Medidas de Dispersão: Variância e Desvio Padrão

### Variância
> É a distância de cada valor do conjunto em relação ao valor central. É A média das diferenças quadradas da média. $σ^2 = (11.56 + 0.16 + 0.36 + 0.16 + 12.96)/5 = 5.4$ 

### Desvio Padrão
> É o que identifica o erro do conjunto de dados, indica o grau de variação. É a raiz quadrada da Variância $σ = \sqrt{(11.56 + 0.16 + 0.36 + 0.16 + 12.96)/5} = 2.24$

### Population vs Sample
> População (population) é o conjunto de todos os dados e sample(amostra) é apenas uma faixa desse conjunto. Quando se trata de amostra, calcula-se a variância amostral. $ S^2 = (11.56 + 0.16 + 0.36 + 0.16 + 12.96)/5 = 6.3 $

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

dados = np.random.normal(200.0, 50.0, 2000)  # 2000 números aleatorios centrados em 200 com desvio padrão de 50 

plt.hist(dados, 50) # histograma com 50 bins
plt.show()
```

```python
dados.std() # Desvio Padrão
```

```python
dados.var() # Variância
```

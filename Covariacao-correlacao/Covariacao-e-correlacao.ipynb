# Covariância e correlação

Covariância mede como duas variaveis se distanciam de suas médias, enquanto a correlação divide a covariância pelos desvios padrões.

```python
%matplotlib inline

import numpy as np
from pylab import *

# Função para calcular todas as médias do dataset 
def de_mean(x):
    xmean = mean(x)
    return [xi - xmean for xi in x]

# Função para calcular covariancia
def covariance(x, y):
    n = len(x)
    return dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

scatter(pageSpeeds, purchaseAmount)

covariance (pageSpeeds, purchaseAmount)
```

```python
# Definindo purchaseAmount em função de pageSpeed
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

scatter(pageSpeeds, purchaseAmount) # Gráfico de dispersão entre pagespeed e purchaseAmount

covariance (pageSpeeds, purchaseAmount) # Covariancia entre pageSpeeds e PurchaseAmount
```

```python
# Definindo função para calcular a correlação
def correlation(x, y):
    stddevx = x.std()
    stddevy = y.std()
    return covariance(x,y) / stddevx / stddevy  

correlation(pageSpeeds, purchaseAmount)
```

```python
np.corrcoef(pageSpeeds, purchaseAmount)
```

```python
purchaseAmount = 100 - pageSpeeds * 3

scatter(pageSpeeds, purchaseAmount)

correlation (pageSpeeds, purchaseAmount)
```

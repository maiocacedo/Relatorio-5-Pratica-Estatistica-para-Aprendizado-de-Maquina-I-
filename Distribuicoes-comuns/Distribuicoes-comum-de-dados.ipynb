# Distribuições de dados
>

```python
import numpy as np
import matplotlib.pyplot as plt 
```

### Distribuição Uniforme 
> Distribuição de probabilidade contínua ou discreta. A probabilidade gerar qualquer ponto em um intervalo contido no espaço amostral é proporcional ao tamanho do intervalo, ou seja, todos os resultados possuem a mesma probabilidade de acontecerem. 

```python
dados = np.random.uniform(-20, 20, 20000) # Distribuição uniforme de -20 a 20 com 20000 números
plt.hist(dados, 100)
plt.show()
```

### Normal/Gaussiana
> é uma distribuição de probabilidade contínua e simétrica em torno de um valor médio que representa o comportamento de um fenômeno natural de forma aleatória. Uma das mais importantes pra estatísca, e podem ser estimados pela media da amostra e pelo seu Desvio Padrão respectivamente. A média, mediana e moda são iguais e seu formato se assemelha com um sino.


```python
from scipy.stats import norm

dados = np.arange(-10,10,0.10) # sequência de valores igualmente espaçados começando em -10 e indo até 10 de 0.10 em 0.10
plt.plot(dados, norm.pdf(dados)) # Gerando um gráfico de distribuição normal
```

```python
media = 18
desvio = 14
dados = np.random.normal(media, desvio, 10000) # Gerando valores aleatorias para distruibuição normal
plt.hist(dados, 50)
plt.show()
```

### Distribuição Binomial
>  de probabilidade discreta, é utilizada para determinar a probabilidade de certa ocorrência acontecer dentro de um evento com sistema fechado e com o número de tentativas limitado. A probabilidade de sucesso em cada tentativa é constante.

```python
from scipy.stats import binom

dados = np.arange(0, 10, 0.001) # sequência de valores igualmente espaçados começando em 0 e indo até 10 de 0.001 em 0.001
plt.plot(dados, binom.pmf(dados, 10, 0.5)) # Gerando gráfico com distribuição binomial
```

### Distribuição de poisson
> é uma distribuição discreta de probabilidade, aplicável a ocorrência de um evento em um intervalo especificado O número médio de eventos é constante, e os eventos ocorrem independentemente entre si.

```python
from scipy.stats import poisson

media = 300
dados = np.arange(200, 400, 0.25) # sequência de valores igualmente espaçados começando em 200 e indo até 400 de 0.25 em 0.25
plt.plot(dados, poisson.pmf(dados, media)) # Gerando um gráfico com distribuição de poisson
```

### Distruibuição Exponencial
> é um distribuição contínua que descreve o tempo entre eventos em um processo de Poisson. É a distribuição que modela o tempo de espera entre eventos independentes. 

```python
from scipy.stats import expon

dados = np.arange(0.1,10,0.10) # sequência de valores igualmente espaçados começando em 0.1 e indo até 10 de 0.1 em 0.1
plt.plot(dados, expon.pdf(dados)) # Gerando gráfico com distribuição exponencial.
```
